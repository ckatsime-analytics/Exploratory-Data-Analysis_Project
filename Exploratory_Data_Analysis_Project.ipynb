##Outliers Idendification on Availability_365

```python
##Box Plot

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['availability_365'])
plt.title('Box Plot of Availability_365')
plt.xlabel('Availability_365')
plt.show()
```

```python
Q1 = df['availability_365'].quantile(0.25)
Q3 = df['availability_365'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers = df[(df['availability_365'] < lower_bound) | (df['availability_365'] > upper_bound)]

print(f"Shape of DataFrame with outliers: {outliers.shape}")
display(outliers.head())
```

##Interpretation

I've identified the outliers in the availability_365 column using the IQR method. It appears there are no outliers based on this calculation, as the shape of the DataFrame with outliers is (0, 22), meaning no rows met the outlier criteria.

# **Bi Variate Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by neighbourhood_group and room_type, then calculate the mean price
grouped_data = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.barplot(x='neighbourhood_group', y='price', hue='room_type', data=grouped_data)
plt.title('Average Price by Neighbourhood Group and Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

# Relationship between Number of Reviews and Price

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=df)
plt.title('Relationship between Number of Reviews and Price')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

The scatter plot shows the relationship between 'number of reviews' and 'price'.

From the plot, it appears that most listings have a relatively low number of reviews (generally less than 200). Within this range, there isn't a clear strong linear correlation between the number of reviews and the price. There are listings with few reviews that have high prices, and listings with many reviews that have moderate prices.

It's possible that properties with very high prices might have fewer reviews due to their exclusivity or lower occupancy rates, while more affordable listings might accumulate more reviews due to higher turnover. However, based purely on this scatter plot, a direct strong positive or negative linear relationship between the two variables is not immediately evident.

#Pairwise Relationships of Price, Minimum Nights, Number of Reviews, and Availability by Room Type

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the relevant columns for the pairplot
columns_to_plot = ['price', 'minimum_nights', 'number_of_reviews', 'availability_365', 'room_type']

# Create a pairplot
plt.figure(figsize=(15, 12))
sns.pairplot(df[columns_to_plot], hue='room_type', diag_kind='kde')
plt.suptitle('Pairwise Relationships of Price, Minimum Nights, Number of Reviews, and Availability by Room Type', y=1.02) # Adjust suptitle position
plt.tight_layout()
plt.show()
```

##Here are some key observations from the pairplot:

Price vs. Other Variables:

Price vs. Minimum Nights: Most listings have a relatively low minimum_nights (around 1 to 30), and within this range, prices vary widely. There doesn't appear to be a strong linear correlation. Some very high-priced listings might have higher minimum nights, but this isn't a general rule.
Price vs. Number of Reviews: As observed previously, the majority of listings have fewer reviews. There's no clear pattern suggesting that more reviews lead to higher or lower prices. High-priced listings can have few or many reviews.
Price vs. Availability_365: Listings with higher availability (closer to 365 days) show a wide range of prices. It doesn't seem to have a direct linear impact on price.
Room Type Differentiation:

Entire home/apt (blue): This room type generally shows a higher price distribution compared to 'Private room' and 'Shared room'. It also tends to have a broader spread across minimum_nights and availability_365.
Private room (orange): These typically fall in the mid-range of prices. They are prevalent across various number_of_reviews and availability_365 values.
Shared room (green) and Hotel room (red): These are less frequent and generally occupy the lower end of the price spectrum. 'Hotel rooms' seem to be concentrated in specific areas or price ranges.
Diagonal Histograms/KDE Plots: The diagonal plots show the distribution of each variable. For instance, the price distribution is skewed right, indicating many lower-priced listings and fewer very high-priced ones. Similar patterns are seen for minimum_nights and number_of_reviews.

Overall, the room_type is a significant factor in differentiating pricing and other characteristics, with 'Entire home/apt' commanding higher prices. The other numerical variables (minimum_nights, number_of_reviews, availability_365) do not exhibit strong direct linear relationships with price, but their distributions and interactions with room_type provide context about the rental market.



# Geographical Distribution of Listings by Room Type

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.scatterplot(x='longitude', y='latitude', hue='room_type', data=df, s=20, alpha=0.7)
plt.title('Geographical Distribution of Listings by Room Type')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Room Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The plot clearly shows the geographical spread of rental listings across New York. We can see distinct clusters corresponding to different boroughs.
Manhattan (the dense cluster on the right) appears to have a high concentration of all room types, particularly 'Entire home/apt' (blue) and 'Private room' (orange). The high density in Manhattan suggests it's a popular area for all types of rentals.
Brooklyn (the cluster to the left of Manhattan) also shows a significant number of listings, with 'Entire home/apt' and 'Private room' being dominant.
Queens (further east from Manhattan/Brooklyn) and The Bronx (north of Manhattan) show listings spread out, with 'Private room' and 'Entire home/apt' still being prevalent, but generally less dense than Manhattan or Brooklyn.
Staten Island (the southwestern cluster) has fewer listings compared to other boroughs, primarily showing 'Private room' and 'Entire home/apt' options.
'Hotel room' (red dots) and 'Shared room' (green dots) are less frequent overall and are scattered, with 'Hotel room' listings often appearing in central, tourist-heavy areas like Manhattan.
In general, the plot provides a good visual summary of where different types of accommodation are located, highlighting the concentration of rentals in Manhattan and Brooklyn.

#Correlation Heatmap of Numerical Columns

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

 # Interpretation of what we can observe:

Strong Positive Correlations (closer to 1):

price and price_per_bed have a high positive correlation (0.69). This is expected, as price_per_bed is derived from price.
beds and price show a moderate positive correlation (0.42), indicating that properties with more beds tend to have higher prices.
Weak or Negligible Correlations (closer to 0):

Most other numerical columns like latitude, reviews_per_month, calculated_host_listings_count, number_of_reviews_ltm, number_of_reviews, minimum_nights, and availability_365 show very weak correlations with price.
For instance, number_of_reviews and price have a very weak negative correlation (-0.04), confirming our earlier observation from the scatter plot that there isn't a strong linear relationship.
Negative Correlations (closer to -1):

longitude has a weak negative correlation with price (-0.19). This suggests that as you move east (increasing longitude), prices tend to decrease slightly, which aligns with geographical patterns in New York City where areas like Manhattan (further west) often have higher prices than some eastern boroughs.
Other Relationships:

number_of_reviews and reviews_per_month show a strong positive correlation (0.67), which is logical as more reviews generally lead to a higher rate of reviews per month.
calculated_host_listings_count has a moderate correlation with number_of_reviews (0.24), implying that hosts with more listings might accumulate more reviews overall.
In summary, the heatmap visually reinforces that price_per_bed and beds are the most positively correlated features with price, while longitude shows a notable, albeit weak, negative correlation. Most other numerical features have very little linear relationship with the price column.
